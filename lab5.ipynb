## Parte 1: Preparación de Datos

**Importación de Datos**

```python
df = pd.read_csv('UK_foods.csv')
df.rename(columns={'Unnamed: 0': 'Food'}, inplace=True)
```

```python
df.head()
```

**Exploración Estadística**

```python
from ydata_profiling import ProfileReport
profile = ProfileReport(df, title='Food', explorative=True)
profile
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
foods = df['Food']
data = df.drop('Food', axis=1)
```

```python
scaler = MinMaxScaler()
data = scaler.fit_transform(data)
scaled_data  = pd.DataFrame(data, columns=df.columns[1:])
```

```python
scaled_data.head()
```

## Parte 2: Construcción del Autocodificador

```python
import tensorflow as tf
from tensorflow.keras import layers, models
```

**Autocodificador**

```python
input_dim = scaled_data.shape[1]
encoding_dim = 2
```

```python
input_layer = layers.Input(shape=(input_dim,))
```

```python
encoder = layers.Dense(10, activation='relu')(input_layer)
encoder = layers.Dense(encoding_dim, activation='relu')(encoder)
```

```python
decoder = layers.Dense(10, activation='relu')(encoder)
decoder = layers.Dense(input_dim, activation='sigmoid')(decoder)
```

```python
autoencoder = models.Model(input_layer, decoder)
autoencoder.compile(optimizer='adam', loss='mse')
```

```python
autoencoder.summary()
```

### Training

```python
autoencoder.fit(scaled_data, scaled_data, epochs=50, batch_size=4, shuffle=True)
```

```python
loss = autoencoder.evaluate(scaled_data, scaled_data)
print(f'Pérdida: {loss}')
```

## Parte 3: Reducción de Dimensionalidad y Visualización

```python
encoded_part = models.Model(input_layer, encoder)
encoded_data = encoded_part.predict(scaled_data)
```

```python
import matplotlib.pyplot as plt

plt.scatter(encoded_data[:, 0], encoded_data[:, 1], c='blue', s=10)

for i, food in enumerate(foods):
    plt.text(encoded_data[i, 0], encoded_data[i, 1], food, rotation=45)

plt.show()
```

Las dimensiones resultantes del proceso de re-dimensionamiento no tienen un significado como tal. Aunque, los pesos indican la importancia de cada variable en la dimensión latente.
